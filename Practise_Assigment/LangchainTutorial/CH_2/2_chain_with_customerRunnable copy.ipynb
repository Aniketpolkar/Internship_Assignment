```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite-preview-09-2025",temperature=0)
```

# **CHIAN WITH CUSTOM RUNNABLE**

```python
# Task-1 [prompt] 
prompt_template = ChatPromptTemplate.from_messages(
    ["system","You are a helpful assistant",
    ("human","{input} give one word answer")
    ])

# Task-2 [LLM]
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite-preview-09-2025",temperature=0)

# Task 3 [String Parser]
str_parser = StrOutputParser()
```

```python

# Task 4 [Custom runnable]
from langchain_core.runnables import RunnableLambda

def dictionary_maker(text:str)->dict:
    return {"text":text}

dictionary_maker_runnable = RunnableLambda(dictionary_maker)    
```

```python
# Task 5 [Template for Post]
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import PromptTemplate

prompt_post = ChatPromptTemplate.from_messages(
    messages = [
        ("system","Your are a social media generator"),
        ("human", "Create a post for the following text for Linkdin: {text}")
        ]
)
```

```python
# Task 6 [Generate for Post]


llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite-preview-09-2025",temperature=0)
```

```python
# Task 7 [str parser]

str_parser = StrOutputParser()
```

```python
# Chain 

chain = prompt_template | llm_gemini | str_parser |  dictionary_maker | dictionary_maker_runnable | prompt_post | llm_gemini | str_parser

chain.invoke("What is the captial of Maharashtra? ")
```
